```
import matplotlib.pyplot as plt

# Sample data
months = ["January", "February", "March", "April", "May", "June"]
product_a_sales = [1200, 1500, 1700, 1600, 1800, 2000]
product_b_sales = [800, 950, 1100, 1000, 1300, 1500]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(months, product_a_sales, marker='o', label='Product A Sales', color='blue')
plt.plot(months, product_b_sales, marker='o', label='Product B Sales', color='green')

# Add titles and labels
plt.title('Monthly Sales Data for Product A and B (2025)', fontsize=14)
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```
